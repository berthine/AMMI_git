## Using Logistic regression

```python
class Multi_LogisticRegression:
    
    #Data
    #Initialize weight
    #compute gradient
    # predict with sigmoid function
    #predict class
    
    def __init__(self, nbr_iter = 10000, epsilon = 0.001, lr = 0.001):
        self.nbr_iter = nbr_iter
        self.epsilon = epsilon
        self.lr = lr
    
    def fit(self, X,y):
        self.unique_label = np.unique(y)
        X = self.add_bias(X)
        self.weights = np.zeros((len(self.unique_label), X.shape[1]))
        y = self.one_hot_encode(y)
        
        for i in range(self.nbr_iter):
            predictions = self.predict(X, False)
            gradient = np.dot((y - predictions).T, X)
            update = self.lr*gradient
            self.weights = self.weights + update
            if np.max(np.abs(update))<self.epsilon: break
     
    def one_hot_encode(self, y):
        u_y = list(self.unique_label)
        encoded_y = np.zeros((len(y), len(u_y)))
        for j, i in enumerate(y):
            idx = u_y.index(i)
            encoded_y[j][idx] = 1
        return encoded_y
        
                
    def softmax(self, p):
        return np.exp(p)/np.sum(np.exp(p), axis=1).reshape(-1,1)
    
    def predict(self, X, no_bias=True):
        if no_bias: X=self.add_bias(X)
        probabilities =  np.dot(X, self.weights.T)
        return self.softmax(probabilities)
    
    def pred_classes(self, X):
        self.prob = self.predict(X)
        return np.vectorize(lambda x: self.unique_label[x])(np.argmax(self.prob ,  axis = 1))
        
        
    def add_bias(self, X):
        return np.insert(X, 0, np.ones(X.shape[0]), axis = 1)
    
    def cross_entropy(self, predictions, y):
        return np.abs(np.sum((y*np.log(predictions) + (1-y)*np.log((1-predictions)))))
    
    def accuracy(self, X, y):
        return ("Accuracy: %.2f" % accuracy_score(y, self.pred_classes(X)))
        
```

```python
 
class GDA:
    
    def __init__(self):
        self.n = None
        self.unique_target = None
        self.m =  None
            
        
    def fit(self, X, y):
         
        self.n = len(y)
        self.unique_target = np.unique(y)
        self.m =  X.shape[1]
        
        # Compute Mean
        self.mean_vectors = []
        for cl in self.unique_target:
            self.mean_vectors.append(np.mean(X[y==cl], axis=0).reshape(-1,1))
        
        #Compute Phi
        self.list_phi=[]
        
        for class_k in self.unique_target :
            resp_class = np.sum(y==class_k)/self.n
            self.list_phi.append(resp_class)
            
        #Compute matrix Sigma    
        sigma = 0
        for i in range(X.shape[0]):
            #print(self.mean_vectors)       
            sigma += np.dot((X[i].reshape(-1,1) - self.mean_vectors[y[i]]),(X[i].reshape(-1,1)- self.mean_vectors[y[i]]).T)
        self.sigma_matrix=sigma/self.n
        
      # Compute probabiities
    
    def compute_p_X(self, X):
        P_X = []
        det_sigma = np.linalg.det(self.sigma_matrix)
        if det_sigma == 0:
            self.sigma_matrix = self.sigma_matrix + 0.01*np.eye(self.m)
            inv_sigma = np.linalg.inv(self.sigma_matrix)
            det_sigma = np.linalg.det(self.sigma_matrix)
            
            for i in range(len(self.unique_target)):
                P_X.append( (2*np.pi)**(-self.m/2) * (det_sigma)**(-0.5)*\
                    np.exp(np.dot(np.dot(-0.5*(X - self.mean_vectors[i]).T, inv_sigma), (X - self.mean_vectors[i]))))
        else:
            
            inv_sigma = np.linalg.inv(self.sigma_matrix)
            
            for i in range(len(self.unique_target)):
                P_X.append( (2*np.pi)**(-self.m/2) * (det_sigma)**(-0.5)*\
                    np.exp(np.dot(np.dot(-0.5*(X - self.mean_vectors[i]).T,inv_sigma), (X - self.mean_vectors[i]))))
        
        return P_X
    # predict
    def predict_Y(self, X):
        
        prediction = []
        for i in range(len(X)):
            probs=[]
            for k in range(len(self.unique_target)):
                probs.append(self.compute_p_X(X[i].reshape(-1,1))[k] * self.list_phi[k]) #probability of y given x when y=0, P(y=0|x),
                               
            prediction.append(np.argmax(probs))
            
        self.predict=np.array(prediction)
        return self.predict

    # Accuracy
    def accuracy(self, x,y):
        return ("Accuracy: %.2f" % accuracy_score(y, self.predict_Y(x)))
    


    
```

```python
class Naive_Bayes:
    
    def __init__(self):
        self.n = None
        self.classes = None
        self.m =  None
        self.predictions = None
        
    def fit(self, X, y):
        self.n = len(y)
        self.classes = sorted(np.unique(y))
        self.m =  X.shape[1]
        
        
        self.phi_y = np.zeros(len(self.classes))
        for i in range(len(self.classes)):
            self.phi_y[i] = (np.sum(y==i))/np.sum(len(y))
        
        
        self.phi_x = np.zeros((self.m,len(self.classes)))
        for i in range(0, self.m):
            for c in self.classes:
                self.phi_x[i, c] = ((np.sum((X[:,i]==1)*(y==c)))+1)/(np.sum(y==c) + self.m)
                
    #Predict
    
    def predict(self, x):
        self.result = np.zeros((x.shape[0],len(self.classes)))
        for i in range(x.shape[0]):
            for c in self.classes:
                X = x[i]
                phi_x_c = self.phi_x[:,c]
                phi_x_y = np.zeros_like(X,dtype=np.float64)
                phi_x_y[X==1] = phi_x_c[X==1]
                phi_x_y[X==0] = (1 - phi_x_c)[X==0]
                phi_y_c = self.phi_y[c]
                self.result[i,c] = np.sum(np.log(phi_x_y)) + np.log(phi_y_c)
                
                
        self.predictions=np.argmax(self.result,axis=1)
        
        return self.predictions
        
    #Accuracy 
    
    def accuracy(self, x, y):
        return ("Accuracy: %.2f" % accuracy_score(y, self.predict(x)))
    
    
    
    
```
